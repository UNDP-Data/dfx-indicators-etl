# World Bank: Indicators API

Retrieving data from World Bank's [Indicator API](https://datahelpdesk.worldbank.org/knowledgebase/topics/125589-developer-information).

## Libraries

```python
from dfpp.transformation.sources.world_bank import Connector
```

## Data Collection

```python
connector = Connector(connections=5)
```

### Indicator Metadata

```python
df_indicators = await connector.get_indicators()
print("Shape:", df_indicators.shape)
display(df_indicators.head())
```

```python
df_indicators.nunique()
```

### Indicator Series

```python
df_series = await connector.get_series("SP.POP.TOTL")
print("Shape:", df_series.shape)
display(df_series.head())
```

```python
df_series.duplicated(subset=["countryiso3code", "date"]).sum()
```
