```python
async with StorageManager() as storage_manager:
    source_file_name = os.path.join(
        storage_manager.sources_path,
        f"{source_cfg['id'].upper()}.{source_cfg['file_format']}",
    )

    data = await storage_manager.read_blob(path=source_file_name)
```

```python
df_source = df_source.loc[
    df_source[source_cfg["datetime_column"]].str.contains(r"^\d{4}$")
].reset_index(drop=True)
df_source.rename(
    columns={
        source_cfg["datetime_column"]: "year",
        source_cfg["country_iso3_column"]: "alpha_3_code",
        indicator_cfg["value_column"]: indicator_cfg["indicator_id"],
    },
    inplace=True,
)
df_source["year"] = df_source["year"].astype("int")
```

```python
assert df_source["FREQ"].value_counts().shape[0] == 1
```

```python
df_source = df_source.loc[df_source["FREQ"] == "A"]
```

```python
df_source.shape
```

```python
df_source["year"].value_counts(dropna=False)
```

```python
df_source["country_or_area"] = cc.pandas_convert(
    series=df_source["alpha_3_code"], src="ISO3", to="name_official"
)
```

```python
df_source["alpha_3_code"].value_counts(dropna=True)
```

```python
country_names = (
    df_source.groupby(["alpha_3_code"])
    .agg(country_or_area=("country_or_area", "first"))
    .reset_index()
)
```

```python
assert country_names.country_or_area.isna().sum() == 0
```

```python
filters = {
    "SEX": indicator_cfg["filter_sex_column"],
    "AGE": indicator_cfg["filter_age_column"],
    "ECO": indicator_cfg["filter_eco_column"],
    "STE": indicator_cfg["filter_ste_column"],
    "SOC": indicator_cfg["filter_value_column"],
}
```

```python
filters
```

```python
valid_filters = {
    key: value
    for key, value in filters.items()
    if key in df_source.columns and value is not None
}
```

```python
valid_filters
```

```python
if valid_filters:
    df_source = df_source.loc[
        (df_source[valid_filters.keys()] == valid_filters.values()).all(axis=1)
    ]
```

```python
df_source = df_source[
    ["alpha_3_code", "country_or_area", indicator_cfg["indicator_id"]]
]
```

```python
df_source.head()
```

```python
async with AsyncPGClient() as pg_client:
    await pg_client.insert_indicator(indicator_cfg["indicator_id"], df_source)
```
