```python
import requests
import pandas as pd

def list_indicators():
    url = "https://www.imf.org/external/datamapper/api/v1/indicators"
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()
    indicators = data.get('indicators', {})
    
    indicator_list = []
    
    for key, value in indicators.items():
        value['id'] = key
        indicator_list.append(value)
    
    df = pd.DataFrame(indicator_list)
    return df
```

```python
df_indicators = list_indicators()
```

```python
def get_imf_indicator_data(
    indicator_id: str,
):
    indicator_url = "https://www.imf.org/external/datamapper/api/v1"

    response = requests.get(
        url=f"{indicator_url}/{indicator_id}"
    )
    response_body = json.loads(response.text)
    response_values = response_body.get("values")

    if not response_values:
        return pd.DataFrame()

    indicator_df = pd.DataFrame.from_records(
        response_body["values"][indicator_id]
    ).sort_index()

    return indicator_df
```
